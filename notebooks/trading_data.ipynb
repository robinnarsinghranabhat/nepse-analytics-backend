```python
MAX_LISTED_SHARE = 1000000000 ## 1 Billion.
from app.data_core.constants import SECTORS

from dataclasses import dataclass
@dataclass
class FilterParam:
    min_val : int
    max_val: int
    sort: str = ""


from typing import Union
def get_data_by_params(
        df,
        sectors=[    "Microfinance",
    "Commercial Banks",
    "Hydro Power"], 
        start_date:str="2023-01-01",
        end_date: str="2024-01-01",
        top_n:int=3,
        sort_by: Union[str | list]=[],
        listed_share: [0, MAX_LISTED_SHARE, false]
        max_listed_share=MAX_LISTED_SHARE,
        min_listed_share=1000,
        company_names=["NABIL", "API"], 
        ):
    
    """
    `top_n` : Choose only first 3 stocks from each sector
    `sort_by` : How to sort the top_n from each sector.
        Options are -> ['']
    """
    if top_n and top_n > 0:
        assert sort_by is not None, "Provide params to sort at sector level"
    df = df[ (df['sector'].isin(sectors)) & (df.date_scraped >= start_date) 
            & (df.date_scraped <= end_date) & (df.stockListedShares >= min_listed_share)
            & (df.stockListedShares <= max_listed_share)]
    

    


    return 
```
